```python
N,M=map(int,input().split())
E=[[0]*(N+1) for _ in range(N+1)]
for _ in range(M):
    a,b,c=map(int,input().split())
    E[a][b]=c
    E[b][a]=c   
ans=0
used=[False]*(N+1)

def dfs(v,s):
    global ans
    used[v]=True
    if s>ans: 
        ans=s
    for i in range(1,N+1):
        if not used[i] and E[v][i]:
            dfs(i,s+E[v][i])
    used[v]=False

for i in range(1,N+1):
    dfs(i,0)
print(ans)
```

```python
def dfs(graph, node, visited):
    if visited[node]:
        return
    
    visited[node] = True
    print(f"Visiting node {node}")
    
    for neighbor in graph[node]:
        print(f"  Exploring neighbor {neighbor} of node {node}")
        dfs(graph, neighbor, visited)

# グラフの隣接リスト表現
graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0, 5],
    3: [1],
    4: [1, 6],
    5: [2],
    6: [4]
}

# 初期化
visited = [False] * len(graph)

# 始点からDFSを開始
dfs(graph, 0, visited)
```
