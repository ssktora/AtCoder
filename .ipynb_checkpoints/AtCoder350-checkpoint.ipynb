```python
ans_list = []
for i in range(1,350):
    if i == 316:
        continue
    p = str(i)
    if len(p) == 1:
        ans_list.append("ABC00"+p)
    elif len(p) == 2:
        ans_list.append("ABC0"+p)
    else:
        ans_list.append("ABC"+p)

#print(len(ans_list))
s = input()
if s in ans_list:
    print("Yes")
else:
    print("No")
```

```python
N, Q = map(int, input().split())
T = list(map(int, input().split()))
check = [True for i in range(N)]
#print(check)
for p in range(Q):
    #print(check)
    i = T[p]
    if check[i-1]:
        check[i-1] = False
    else:
        check[i-1] = True
        
ans = check.count(True)
print(ans)
```

```python
N = int(input())
A = list(map(int, input().split()))
P = []
for i in range(len(A)):
    a = A[i]
    P.append((a,i))
check = [False for i in range(N)]
P = sorted(P)
# print(P)
ans = 0
ans_list = []
frag = True
while frag:
    frag = False
    for j in range(len(P)):
#         print(P)
#         print(ans_list)
#         print("--------")
        a, i = P[j][0], P[j][1]
        if i != j:
            P[i], P[j] = (a,i), (P[i][0], P[i][1])
            ans += 1
            ans_list.append((a, P[j][0]))
            frag = True

#print(P)
print(ans)
for d in ans_list:
    if d[0] < d[1]:
        print(d[0], d[1])
    else:
        print(d[1], d[0])
    
    
    
```

```python
N, M = map(int, input().split())
from collections import defaultdict, Counter
friends = {}
for user in range(1,N+1):
    friends[user] = [False for i in range(N)]
    friends[user][user-1] = "-"
for i in range(M):
    x, y = map(int, input().split())
    friends[x][y-1] = True
    friends[y][x-1] = True
    
from itertools import combinations

people = [i for i in range(1,N+1)]
# 3人を選ぶ組み合わせをすべて列挙し、リストに格納する
combinations_list = list(combinations(people, 3))

# 組み合わせを表示する
for combo in combinations_list:
    x = combo[0]
    y = combo[1]
    z = combo[2]
    if friends[x][y] and not friends[x][z]:
        ans += 1
    if friends[y][z] and not friends[y][x]:
        ans += 1
    if friends[z][x] and not friends[z][y]:
        ans += 1

# ans = 0
# print(friends)
# for user, f_list in friends.items():
    
#     for f in f_list:
#         counter1 = Counter(friends[f])
#         counter2 = Counter(friends[user])
#         difference = counter1 - counter2
#         #print(difference)
#         count = sum(difference.values())
#         ans += (count-1)
#         for k in difference.keys():
#             if k != user:
#                 friends[k].append(user)
#                 friends[user].append(k)
#         print(friends)
        
        
print(ans)
    
```
