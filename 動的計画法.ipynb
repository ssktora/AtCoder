# A-Frog1

```python
N = int(input())
h_list = list(map(int, input().split()))
dp = [0 for _ in range(N)]
dp[1] = abs(h_list[0]-h_list[1])
if N >= 3:
    for i in range(2,N):
        a = dp[i-2] + abs(h_list[i-2]-h_list[i])
        b = dp[i-1] + abs(h_list[i-1]-h_list[i])
        dp[i] = min(a,b)
    
    
print(dp[-1])
```

# B-Frog2

```python
N, K = map(int, input().split())
h_list = list(map(int, input().split()))
dp = [0 for _ in range(N)]
dp[1] = abs(h_list[0]-h_list[1])
if N >= 3:
    for i in range(2,N):
        if i <= K:
            ans = 100000000000000
            for s in range(1,i+1):
                p = dp[i-s] + abs(h_list[i-s]-h_list[i])
                ans = min(ans,p)
            dp[i] = ans
        else:
            ans = 1000000000000000
            for t in range(1,K+1):
                p = dp[i-t] + abs(h_list[i-t]-h_list[i])
                ans = min(ans,p)
            dp[i] = ans
    
    
print(dp[-1])
```

# C-Vacation

```python
N = int(input())
dp = [[0,0,0] for _ in range(N)]
a, b, c = map(int, input().split())
dp[0][0] = a
dp[0][1] = b
dp[0][2] = c
for i in range(1,N):
    a, b, c = map(int, input().split())
    dp[i][0] = max(dp[i-1][1]+a,dp[i-1][2]+a)
    dp[i][1] = max(dp[i-1][0]+b,dp[i-1][2]+b)
    dp[i][2] = max(dp[i-1][0]+c,dp[i-1][1]+c)
            
print(max(dp[-1]))
```

# D-Knapsak1

```python
N, W = map(int,input().split())
dp = [[0]*(W+1) for _ in range(N)]
print(dp)
w, v = map(int, input().split())
for j in range(w,W+1):
    dp[0][j] = v

for i in range(1,N):
    w, v = map(int,input().split())
    for j in range(W+1):
        if j - w >= 0:
            dp[i][j] = max(dp[i-1][j-w] + v,dp[i-1][j])
        else:
            dp[i][j] = dp[i-1][j]
            
print(max(dp[-1]))
        
```

# E-Knapsack2

```python
N, W = map(int, input().split())
V = 0
a_list = []
for i in range(N):
    w, v = map(int, input().split())
    V += v
    a_list.append((w,v))
    
INF = 100000000000000000
    
dp = [[INF]*(V+1) for _ in range(N+1)]  
for i in range(N):
    dp[i][0] = 0
    
for i in range(N):
    w = a_list[i][0]
    v = a_list[i][1]
    for j in range(V+1):
        if j-v >= 0:
            dp[i+1][j] = min(dp[i+1][j], dp[i][j-v]+w)
        else:
            dp[i+1][j] = dp[i][j]
            
for ans in range(len(dp[-1])-1,-1,-1):
    if dp[-1][ans] <= W:
        print(ans)
        break
        
            
```
