# 再帰関数

```python
n, x = map(int, input().split())
def rec(N,X,L,S):
    if N == 0:
        return 1
    
    if X == 1:
        print("a")
        return 0
    
    elif X <= L[N-1]+1:
        return rec(N-1,X-1,L,S)
    
    elif X == L[N-1]+2:
        return S[N-1] + 1
    
    elif X <=  L[N-1] *2 + 2:
        return rec(N-1, X - L[N-1] -2, L, S) + S[N-1] + 1
    
    else:
        return 2 * S[N-1] + 1
    
L = [1] * (n+1)
S = [1] * (n+1)
for i in range(1, n+1):
    L[i] = 2 * L[i-1] + 3
    S[i] = 2 * S[i-1] + 1
    

ans = rec(n, x, L, S)
print(ans)
```

# 再帰的な全探索

```python
N = 3
count = 0
def rec(A):
    global count
    print(count)
    count += 1
    if len(A) == N:
        return 
    
    for v in range(2):
        A.append(v)
        print(A)
        rec(A)
        print(A)
        A.pop()
        print(A)
        print("--------------")

rec([])
        
```
