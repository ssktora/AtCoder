```python
H, W, N = map(int, input().split())
def update_iW(i):
    i += 1
    if i == 4 or i == 5:
        i -= 4
        return i
    else:
        return i
    
def update_iB(i):
    i -= 1
    if i == -1 or i == -2:
        i += 4
        return i
    else:
        return i
    
def update_p(p, now_direction):
    if now_direction == "N":
        i, j = p[0]-1, p[1]
    elif now_direction == "E":
        i, j = p[0], p[1]+1
    elif now_direction == "S":
        i, j = p[0]+1, p[1]
    elif now_direction == "W":
        i, j = p[0], p[1]-1
    if j > W:
        j = 1
    if j < 1:
        j = W
    if i > H:
        i = 1
    if i < 1:
        i = H
    return (i, j)
    
        
        
        
direction = ["N", "E", "S", "W"]
brack = []
now = (1, 1)
d = 0
now_direction = direction[d]
for i in range(N):
    if now not in brack:
        brack.append(now)
        d = update_iW(d)
        now_direction = direction[d]
        now = update_p(now, now_direction)
    else:
        brack.remove(now)
        d = update_iB(d)
        now_direction = direction[d]
        now = update_p(now, now_direction)
    print(now_direction)
    print(brack)

ans = []
for i in range(H):
    row = []
    for j in range(W):
        if (i+1,j+1) in brack:
            row.append("#")
        else:
            row.append(".")  
    ans.append(row)
    
for a in ans:
    print("".join(a))
```
