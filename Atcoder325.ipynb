```python
from queue import Queue
H, W = map(int, input().split())
S = [list(input()) for _ in range(H)]
dx = [0,1,1,1,0,-1,-1,-1]
dy = [-1,-1,0,1,1,1,0,-1]
used = [[False]*W for _ in range(H)]
ans = 0
for i in range(H):
    for j in range(W):
        if S[i][j] == "." or used[i][j]:
            continue
        
        que = Queue()
        que.put((i,j))
        while not que.empty():
            x,y = que.get()
            for d in range(8):
                nx = x + dx[d]
                ny = y + dy[d]
                if 0 <= nx < H and 0 <= ny < W and S[nx][ny] == "#" and not used[nx][ny]:
                    que.put((nx,ny))
                    used[nx][ny] = True
        ans += 1
        
print(ans)
        
```
