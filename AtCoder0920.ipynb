```python
a, b = map(int, input().split())
t = a
n_list = []
while t != b+1:
    n_list.append(t)
    t += 1

ans = 1
#print(n_list)
for d in n_list:
    ans *= d
    #print(ans)
    
if ans > 0:
    print("Positive")
elif ans < 0:
    print("Negative")
else:
    print("Zero")
    
```

```python
v = list(map(int, input().split()))
    
odd = sum(1 for i in v if i % 2 != 0)
even = sum(1 for i in v if i % 2 == 0)
    
ans = 0

if odd == 2:
    ans += 1
    v = [i + 1 if i % 2 == 1 else i for i in v]
elif even == 2:
    ans += 1
    v = [i + 1 if i % 2 == 0 else i for i in v]

ma = max(v)
ans += sum((ma-i)//2 for i in v)
print(ans)
```

```python
def solve(N2, N3, N4):
    N1, N2, N3 = N2, N4, N3 // 2
    print(N1, N2, N3)
    ANS = 0

    comb = [(0, 1, 1), (2, 0, 1), (1, 2, 0), (3, 1, 0), (5, 0, 0)]
    for a, b, c in comb:
        k = N1 + N2 + N3
        print("before", k)
        if a > 0:
            k = min(k, N1 // a)
        if b > 0:
            k = min(k, N2 // b)
        if c > 0:
            k = min(k, N3 // c)
        print("after", k)
        ANS += k
        print("ANS", ANS)
        N1 -= a * k
        N2 -= b * k
        N3 -= c * k
        print(N1, N2, N3)

    print("ANS:", ANS)  # Debugging: Print the value of ANS after the loop
    print("N1, N2, N3 after loop:", N1, N2, N3)  # Debugging: Print the values of N1, N2, N3 after the loop

T = int(input())
for _ in range(T):
    print("Test case:", _ + 1)  # Debugging: Print the current test case number
    n2, n3, n4 = map(int, input().split())
    print("Inputs:", n2, n3, n4)  # Debugging: Print the input values for the current test case
    solve(n2, n3, n4)
```
