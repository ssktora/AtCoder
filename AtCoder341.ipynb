```python
def is_valid_move(x, y, grid):
    # グリッドの範囲を超えていないか、壁や障害物がないかを確認
    return 0 <= x < len(grid) and 0 <= y < len(grid[0]) and grid[x][y] != '#'

def find_next_positions(i, j, grid, d):
    if d == "L":
        new_i, new_j = i, j-1
    elif d == "R":
        new_i, new_j = i, j+1
    elif d == "U":
        new_i, new_j = i-1, j
    else:
        new_i, new_j = i+1, j
    if is_valid_move(new_i, new_j, grid):
        return (True, (new_i,new_j))
    else:
        return (False, (i,j))

H, W, N = map(int, input().split())
T = input()
S = [list(input()) for _ in range(H)]
land = []
for i in range(H):
    for j in range(W):
        if S[i][j] == ".":
            land.append((i,j))
ans = 0            
for p in range(len(land)):
    i = land[p][0]
    j = land[p][1]
    frag = True
    for q in range(len(T)):
        #print(i,j)
        d = T[q]
        check = find_next_positions(i, j, S, d)
        if check[0]:
            i = check[1][0]
            j = check[1][1]
        if not check[0]:
            frag = False
            break
    #print("----")
    if frag:
        ans += 1
        
print(ans)
```

```python
N,M,K = map(int, input().split())
import math
def lcm(a, b):
    return (a * b) // math.gcd(a, b)
def f(X, N, M):
    l = lcm(N,M)
    c = (X//N)+(X//M)-2*(X//l)
    return c >= K

def binary_search():
    left = 0
    right = 10**18

    while right-left>1:
        mid = (left + right) // 2

        if f(mid,N,M):
            right = mid
        else:
            left = mid
        
    return right

ans = binary_search()
print(ans)

```
