```python
N, T, M = map(int, input().split())
G = {i: set() for i in range(1, N+1)}
for _ in range(M):
    a, b = map(int, input().split())
    G[a].add(b)
    G[b].add(a)

def dfs(nx, mem, emp):
    global ans
    global C

    print("Entering dfs:", nx, mem, emp, C)  # 追加：再帰呼び出し前のデバッグ情報

    if nx == N+1:
        if C[T] > 0:
            ans += 1
            print("Found valid coloring:", nx, mem, emp, C)  # 追加：条件を満たす塗り分けを見つけた場合のデバッグ情報
        return

    for col in range(1, emp+1):
        mem[nx] = col
        C[col] += 1
        ng = False
        for x in G[nx]:
            if mem[x] == col:
                ng = True
        if not ng:
            if col == emp:
                dfs(nx+1, mem, min(T, emp+1))
            else:
                dfs(nx+1, mem, emp)
        mem[nx] = -1
        C[col] -= 1

    print("Exiting dfs:", nx, mem, emp, C)  # 追加：再帰呼び出し後のデバッグ情報
mem = [-1] * (N+1)
mem[1] = 1
ans = 0
C = [0] * (T+1)
C[1] = 1
dfs(2,mem,min(T,2))
print(ans)
```
